```python
# Melhor Série Cômica

import requests
from bs4 import BeautifulSoup

planilha = []

links_episodios = ["https://www.imdb.com/title/tt8270790/?ref_=ttep_ep1",
                   "https://www.imdb.com/title/tt8884332/?ref_=ttep_ep2",
                   "https://www.imdb.com/title/tt8940996/?ref_=ttep_ep3",
                   "https://www.imdb.com/title/tt8956330/?ref_=ttep_ep4",
                   "https://www.imdb.com/title/tt8956336/?ref_=ttep_ep6",
                   "https://www.imdb.com/title/tt8956340/?ref_=ttep_ep7",
                   "https://www.imdb.com/title/tt8890392/?ref_=ttep_ep8",
                   "https://www.imdb.com/title/tt6733136/?ref_=ttep_ep1",
                   "https://www.imdb.com/title/tt9679520/?ref_=ttep_ep2",
                   "https://www.imdb.com/title/tt9679522/?ref_=ttep_ep3",
                   "https://www.imdb.com/title/tt9679524/?ref_=ttep_ep4",
                   "https://www.imdb.com/title/tt9679526/?ref_=ttep_ep5",
                   "https://www.imdb.com/title/tt6733140/?ref_=ttep_ep6",
                   "https://www.imdb.com/title/tt7664192/?ref_=ttep_ep1",
                   "https://www.imdb.com/title/tt8601200/?ref_=ttep_ep2",
                   "https://www.imdb.com/title/tt8601206/?ref_=ttep_ep3",
                   "https://www.imdb.com/title/tt8601210/?ref_=ttep_ep4",
                   "https://www.imdb.com/title/tt8534442/?ref_=ttep_ep5",
                   "https://www.imdb.com/title/tt8564998/?ref_=ttep_ep6",
                   "https://www.imdb.com/title/tt8488908/?ref_=ttep_ep7",
                   "https://www.imdb.com/title/tt8601212/?ref_=ttep_ep8",
                   "https://www.imdb.com/title/tt8578788/?ref_=ttep_ep9",
                   "https://www.imdb.com/title/tt8601222/?ref_=ttep_ep10",
                   "https://www.imdb.com/title/tt8601216/?ref_=ttep_ep11",
                   "https://www.imdb.com/title/tt7853310/?ref_=ttep_ep12",
                   "https://www.imdb.com/title/tt7667344/?ref_=ttep_ep1",
                   "https://www.imdb.com/title/tt7690008/?ref_=ttep_ep2",
                   "https://www.imdb.com/title/tt7772296/?ref_=ttep_ep3",
                   "https://www.imdb.com/title/tt7772302/?ref_=ttep_ep4",
                   "https://www.imdb.com/title/tt7775670/?ref_=ttep_ep5",
                   "https://www.imdb.com/title/tt7775674/?ref_=ttep_ep6",
                   "https://www.imdb.com/title/tt7775676/?ref_=ttep_ep7",
                   "https://www.imdb.com/title/tt7667348/?ref_=ttep_ep8",
                   "https://www.imdb.com/title/tt7978458/?ref_=ttep_ep9",
                   "https://www.imdb.com/title/tt7978462/?ref_=ttep_ep10",
                   "https://www.imdb.com/title/tt7526214/?ref_=ttep_ep1",
                   "https://www.imdb.com/title/tt8608804/?ref_=ttep_ep2",
                   "https://www.imdb.com/title/tt8709138/?ref_=ttep_ep3",
                   "https://www.imdb.com/title/tt8577836/?ref_=ttep_ep4",
                   "https://www.imdb.com/title/tt8526428/?ref_=ttep_ep5",
                   "https://www.imdb.com/title/tt8608802/?ref_=ttep_ep6",
                   "https://www.imdb.com/title/tt8709140/?ref_=ttep_ep7",
                   "https://www.imdb.com/title/tt8566574/?ref_=ttep_ep8",
                   "https://www.imdb.com/title/tt6932620/?ref_=ttep_ep1",
                   "https://www.imdb.com/title/tt7806352/?ref_=ttep_ep2",
                   "https://www.imdb.com/title/tt7870754/?ref_=ttep_ep3",
                   "https://www.imdb.com/title/tt7870756/?ref_=ttep_ep4",
                   "https://www.imdb.com/title/tt7870758/?ref_=ttep_ep5",
                   "https://www.imdb.com/title/tt7870760/?ref_=ttep_ep6",
                   "https://www.imdb.com/title/tt7870762/?ref_=ttep_ep7",
                   "https://www.imdb.com/title/tt9203528/?ref_=ttep_ep1",
                   "https://www.imdb.com/title/tt9525916/?ref_=ttep_ep2",
                   "https://www.imdb.com/title/tt9599994/?ref_=ttep_ep3",
                   "https://www.imdb.com/title/tt9639696/?ref_=ttep_ep4",
                   "https://www.imdb.com/title/tt9663044/?ref_=ttep_ep5",
                   "https://www.imdb.com/title/tt9728350/?ref_=ttep_ep6",
                   "https://www.imdb.com/title/tt9765702/?ref_=ttep_ep7",
                   "https://www.imdb.com/title/tt9799628/?ref_=ttep_ep8",
                   "https://www.imdb.com/title/tt9809498/?ref_=ttep_ep9",
                   "https://www.imdb.com/title/tt9809502/?ref_=ttep_ep10",
                   "https://www.imdb.com/title/tt9809506/?ref_=ttep_ep11",
                   "https://www.imdb.com/title/tt9870384/?ref_=ttep_ep12",
                   "https://www.imdb.com/title/tt9870386/?ref_=ttep_ep13",
                   "https://www.imdb.com/title/tt10005284/?ref_=ttep_ep14"]

for link in links_episodios: # Pega cada link da lista de links para fazer a requisição
    requisicao = requests.get(link)

    sopa = BeautifulSoup(requisicao.content, "html.parser") # Dados brutos da página inteira
    
    # Conjunto que acha o nome da série
    acha_nome = sopa.find("div", {"class": "titleParent"})
    nome_serie = acha_nome.find("a").text

    # Conjunto que acha a temporada e o número do episódio
    acha_temp_episodio = sopa.find("div", {"class": "bp_description"})
    temp_episodio = acha_temp_episodio.find("div", {"class": "bp_heading"}).text.strip() # Temporada e episódio dentro da mesma tag separados por |
    lista = temp_episodio.split(" | ") # Comando split e cria uma lista com os dois itens
    temporada = lista[0] # Primeiro item da lista corresponde à temporada
    episodio = lista[1] # Segundo item da lista corresponde ao episódio
    temporada = temporada.replace("Season ", "") # Limpa para deixar só o número
    episodio = episodio.replace("Episode", "") # Limpa para deixar só o número
    
    # Conjunto que acha o nome do episódio
    acha_nome_episodio = sopa.find("div", {"class": "title_wrapper"}) # Div com o nome do episódio
    nome_episodio = acha_nome_episodio.find("h1").text.strip() # Limpa só o nome do episódio

    # Conjunto que acha a avaliação do episódio
    avaliacao = sopa.find("div", {"class": "imdbRating"})
    nota = avaliacao.find("span", {"itemprop": "ratingValue"}).text.strip()
    votos = avaliacao.find("span", {"itemprop": "ratingCount"}).text.strip()
    votos = votos.replace(",", "") # Limpa para poder converter para inteiro
    
    dicionario = {"serie": nome_serie, "temporada": temporada, "episodio": episodio, "nome_episodio": nome_episodio, "nota": float(nota), "votos": int(votos)} # Monta o dicionário
    planilha.append(dicionario) # Adiciona planilha ao dicionário
    
print(planilha)
```

```python
import pandas

df = pandas.DataFrame(planilha)
df.to_csv("serie-comica.csv", sep = ";", encoding = "utf-8")
df.sort_values(by = "nota", ascending = False)
```
