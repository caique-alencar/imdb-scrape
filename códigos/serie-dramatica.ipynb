```python
# Melhor Série Dramática

import requests
from bs4 import BeautifulSoup

planilha = []

links_episodios = ["https://www.imdb.com/title/tt7073996/?ref_=ttep_ep1",
                   "https://www.imdb.com/title/tt7073998/?ref_=ttep_ep2",
                   "https://www.imdb.com/title/tt7074002/?ref_=ttep_ep3",
                   "https://www.imdb.com/title/tt7074006/?ref_=ttep_ep4",
                   "https://www.imdb.com/title/tt7074010/?ref_=ttep_ep5",
                   "https://www.imdb.com/title/tt7074012/?ref_=ttep_ep6",
                   "https://www.imdb.com/title/tt7074020/?ref_=ttep_ep7",
                   "https://www.imdb.com/title/tt7074022/?ref_=ttep_ep8",
                   "https://www.imdb.com/title/tt7074032/?ref_=ttep_ep9",
                   "https://www.imdb.com/title/tt7074030/?ref_=ttep_ep10",
                   "https://www.imdb.com/title/tt7517316/?ref_=ttep_ep1",
                   "https://www.imdb.com/title/tt7658640/?ref_=ttep_ep2",
                   "https://www.imdb.com/title/tt7658622/?ref_=ttep_ep3",
                   "https://www.imdb.com/title/tt7658626/?ref_=ttep_ep4",
                   "https://www.imdb.com/title/tt7658630/?ref_=ttep_ep5",
                   "https://www.imdb.com/title/tt7517318/?ref_=ttep_ep6",
                   "https://www.imdb.com/title/tt5924366/?ref_=ttep_ep1",
                   "https://www.imdb.com/title/tt6027908/?ref_=ttep_ep2",
                   "https://www.imdb.com/title/tt6027912/?ref_=ttep_ep3",
                   "https://www.imdb.com/title/tt6027914/?ref_=ttep_ep4",
                   "https://www.imdb.com/title/tt6027916/?ref_=ttep_ep5",
                   "https://www.imdb.com/title/tt6027920/?ref_=ttep_ep6",
                   "https://www.imdb.com/title/tt8241576/?ref_=ttep_ep1",
                   "https://www.imdb.com/title/tt9264262/?ref_=ttep_ep2",
                   "https://www.imdb.com/title/tt9264276/?ref_=ttep_ep3",
                   "https://www.imdb.com/title/tt9264282/?ref_=ttep_ep4",
                   "https://www.imdb.com/title/tt9264290/?ref_=ttep_ep5",
                   "https://www.imdb.com/title/tt9297972/?ref_=ttep_ep6",
                   "https://www.imdb.com/title/tt9297976/?ref_=ttep_ep7",
                   "https://www.imdb.com/title/tt9297978/?ref_=ttep_ep8",
                   "https://www.imdb.com/title/tt7258794/?ref_=ttep_ep1",
                   "https://www.imdb.com/title/tt7757530/?ref_=ttep_ep2",
                   "https://www.imdb.com/title/tt8014750/?ref_=ttep_ep3",
                   "https://www.imdb.com/title/tt7831074/?ref_=ttep_ep4",
                   "https://www.imdb.com/title/tt8014756/?ref_=ttep_ep5",
                   "https://www.imdb.com/title/tt7986260/?ref_=ttep_ep6",
                   "https://www.imdb.com/title/tt8014758/?ref_=ttep_ep7",
                   "https://www.imdb.com/title/tt8022752/?ref_=ttep_ep8",
                   "https://www.imdb.com/title/tt8022760/?ref_=ttep_ep9",
                   "https://www.imdb.com/title/tt8022768/?ref_=ttep_ep10",
                   "https://www.imdb.com/title/tt7600128/?ref_=ttep_ep1",
                   "https://www.imdb.com/title/tt7637846/?ref_=ttep_ep2",
                   "https://www.imdb.com/title/tt7793104/?ref_=ttep_ep3",
                   "https://www.imdb.com/title/tt7793106/?ref_=ttep_ep4",
                   "https://www.imdb.com/title/tt7793110/?ref_=ttep_ep5",
                   "https://www.imdb.com/title/tt7793114/?ref_=ttep_ep6",
                   "https://www.imdb.com/title/tt7793116/?ref_=ttep_ep7",
                   "https://www.imdb.com/title/tt7779744/?ref_=ttep_ep8",
                   "https://www.imdb.com/title/tt5791038/?ref_=ttep_ep1",
                   "https://www.imdb.com/title/tt7660884/?ref_=ttep_ep2",
                   "https://www.imdb.com/title/tt7660888/?ref_=ttep_ep3",
                   "https://www.imdb.com/title/tt7660890/?ref_=ttep_ep4",
                   "https://www.imdb.com/title/tt7660892/?ref_=ttep_ep5",
                   "https://www.imdb.com/title/tt7660896/?ref_=ttep_ep6",
                   "https://www.imdb.com/title/tt7660898/?ref_=ttep_ep7",
                   "https://www.imdb.com/title/tt7660900/?ref_=ttep_ep8",
                   "https://www.imdb.com/title/tt7660912/?ref_=ttep_ep9",
                   "https://www.imdb.com/title/tt7660914/?ref_=ttep_ep10",
                   "https://www.imdb.com/title/tt6439874/?ref_=ttep_ep1",
                   "https://www.imdb.com/title/tt6439878/?ref_=ttep_ep2",
                   "https://www.imdb.com/title/tt6439890/?ref_=ttep_ep3",
                   "https://www.imdb.com/title/tt6439900/?ref_=ttep_ep4",
                   "https://www.imdb.com/title/tt6439894/?ref_=ttep_ep5",
                   "https://www.imdb.com/title/tt6439896/?ref_=ttep_ep6",
                   "https://www.imdb.com/title/tt6439898/?ref_=ttep_ep7",
                   "https://www.imdb.com/title/tt6439886/?ref_=ttep_ep8",
                   "https://www.imdb.com/title/tt6439888/?ref_=ttep_ep9",
                   "https://www.imdb.com/title/tt6439892/?ref_=ttep_ep10",
                   "https://www.imdb.com/title/tt6439906/?ref_=ttep_ep11",
                   "https://www.imdb.com/title/tt6439902/?ref_=ttep_ep12",
                   "https://www.imdb.com/title/tt6439910/?ref_=ttep_ep13",
                   "https://www.imdb.com/title/tt6439908/?ref_=ttep_ep14",
                   "https://www.imdb.com/title/tt6439916/?ref_=ttep_ep15",
                   "https://www.imdb.com/title/tt6439912/?ref_=ttep_ep16",
                   "https://www.imdb.com/title/tt6439904/?ref_=ttep_ep17",
                   "https://www.imdb.com/title/tt6439914/?ref_=ttep_ep18"]

for link in links_episodios: # Pega cada link da lista de links para fazer a requisição
    requisicao = requests.get(link)

    sopa = BeautifulSoup(requisicao.content, "html.parser") # Dados brutos da página inteira
    
    # Conjunto que acha o nome da série
    acha_nome = sopa.find("div", {"class": "titleParent"})
    nome_serie = acha_nome.find("a").text

    # Conjunto que acha a temporada e o número do episódio
    acha_temp_episodio = sopa.find("div", {"class": "bp_description"})
    temp_episodio = acha_temp_episodio.find("div", {"class": "bp_heading"}).text.strip() # Temporada e episódio dentro da mesma tag separados por |
    lista = temp_episodio.split(" | ") # Comando split e cria uma lista com os dois itens
    temporada = lista[0] # Primeiro item da lista corresponde à temporada
    episodio = lista[1] # Segundo item da lista corresponde ao episódio
    temporada = temporada.replace("Season ", "") # Limpa para deixar só o número
    episodio = episodio.replace("Episode", "") # Limpa para deixar só o número
    
    # Conjunto que acha o nome do episódio
    acha_nome_episodio = sopa.find("div", {"class": "title_wrapper"}) # Div com o nome do episódio
    nome_episodio = acha_nome_episodio.find("h1").text.strip() # Limpa só o nome do episódio

    # Conjunto que acha a avaliação do episódio
    avaliacao = sopa.find("div", {"class": "imdbRating"})
    nota = avaliacao.find("span", {"itemprop": "ratingValue"}).text.strip()
    votos = avaliacao.find("span", {"itemprop": "ratingCount"}).text.strip()
    votos = votos.replace(",", "") # Limpa para poder converter para inteiro
    
    dicionario = {"serie": nome_serie, "temporada": temporada, "episodio": episodio, "nome_episodio": nome_episodio, "nota": float(nota), "votos": int(votos)} # Monta o dicionário
    planilha.append(dicionario) # Adiciona planilha ao dicionário
    
print(planilha)
```

```python
import pandas

df = pandas.DataFrame(planilha)
df.to_csv("serie-dramatica.csv", sep = ";", encoding = "utf-8")
df.sort_values(by = "nota", ascending = False)
```
